## Estimation of $\pi$ by the Hit-or-Miss method in $[-1,1]^2$

```octave
N = 1000; % The number of samples
```

```octave
X = 2*rand(N,1)-1; Y = 2*rand(N,1)-1;  % The X and Y coordinates of sampled points
```

```octave
I = (X.^2 + Y.^2) < 1;  % Mark the points inside the circle
```

```octave
XI = X(I); YI = Y(I);  % The points inside the circle
```

```octave
XO = X(not(I)); YO = Y(not(I));  % The points outside the circle
```

```octave
plot(cos(0:0.1:2*pi),sin(0:0.1:2*pi),'b');
hold on;
plot(XI,YI,'g.')
plot(XO,YO,'r.')
hold off
```

```octave
Pi_estimate = 4 * mean(I);
Pi_staterror = 4 * std(I) / sqrt(N);
```

```octave
fprintf('Sample size is %5d\n',N)
fprintf('Sample mean is %8.4f\n',Pi_estimate)
fprintf('Stat. error is %8.5f\n',Pi_staterror)
```

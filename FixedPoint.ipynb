# Fixed-Point Method

```octave
function p = plotiter(g, xlim, p0, niter, anim)
%  Plot fixed-point iterations
%  PLOTITER(G,XLIM,X0,NITER,ANIM)
%  G:     name of the function or its reference
%  XLIM:  vector of limits [XMIN XMAX] or [XMIN XMAX YMIN YMAX]
%  P0:    initial value
%  NITER: number of iteration
%  ANIM = TRUE or FALSE: with or without the animation
%
x0 = fzero(@(x) g(x)-x, p0);
%
p = zeros(niter+1,1);
p(1)= p0;
for k=1:niter
    p(k+1)=feval(g,p(k));
end
%
clf;
xx = linspace(xlim(1),xlim(2),100);
plot(xx, xx,'-g');
hold on
plot(xx,g(xx),'-m','LineWidth',3);
plot(xx,xx,'-g','LineWidth',2);
plot( x0, x0, 'o', 'MarkerEdgeColor', 'g','MarkerFaceColor','g', 'MarkerSize',10);
  rx = [p(1) p(1)];
  ry = [p(1) 0];
  plot( rx, ry, '--r');
grid on;
%axes('Position',[xl yl xr yr]);
title(['Iteration k=',int2str(0),'   x_k=',num2str(p(1))], 'FontSize', 18);
plot( p(1), 0, 'o', 'MarkerEdgeColor', 'r','MarkerFaceColor','r', 'MarkerSize',10);
  text( p(1), 0,num2str(0),...
    'VerticalAlignment','bottom',...
    'HorizontalAlignment','center',...
    'FontSize',16)
if anim, pause(1), end
for k=1:niter
  title(['Iteration k=',int2str(k),'   x_k=',num2str(p(k+1))], 'FontSize', 18);
  rx = [p(k) p(k)];
  ry = [p(k) p(k+1)];
  plot( rx, ry, '-b','LineWidth',2);
  if anim, pause(2/niter), end
  rx = [p(k) p(k+1)];
  ry = [p(k+1) p(k+1)];
  plot( rx, ry, '-r','LineWidth',2);
  if anim, pause(2/niter), end
  rx = [p(k+1) p(k+1)];
  ry = [p(k+1) 0];
  plot( rx, ry, '--r');
  plot( p(k+1), 0, 'o', 'MarkerEdgeColor', 'r','MarkerFaceColor','r', 'MarkerSize',10);
  text( p(k+1), 0,num2str(k),...
    'VerticalAlignment','bottom',...
    'HorizontalAlignment','center',...
    'FontSize',16)
  if anim, pause(4/niter), end
end
hold off
end %function
```

## Solve 1 - x^3 = 0 on the interval [0, 1]  -- Diverging iterations

```octave
res1 = plotiter(@(x) 1-x.^3, [0 1], 0.7, 10, 0);
```

## Solve (1 - x).^(1/3) = 0 on the interval [0, 1] -- Converging iterations

```octave
res2 = plotiter(@(x) (1-x).^(1.0/3.0), [0 1], 0.2, 10, 0);
```

## Solve (1+2*x.^3)./(1+3*x.^2) = 0 on the interval [0, 1] - Rapidly converging iterations

```octave
res3 = plotiter(@(x) (1+2*x.^3)./(1+3*x.^2), [0 1], 0.2, 10, 0);
```

## Iteration results for the three functions

```octave
[res1, res2, res3]
```

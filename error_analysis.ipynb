## Hilbert Matrix

```octave
for n=5:14
    A = zeros(n);
    for i=1:n
        for j=1:n
           A(i,j) = 1/(i+j-1);
        end
    end
    c = cond(A,1);  %Use the L1 norm
    d = det(A);
    dd = det(A)*det(inv(A));
    fprintf('n=%2.0f    cond(A)=%e,   det(A)=%e,      det(A)*det(inv(A))=%6.4f\n',n,c,d,dd)
end
```
